## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing dataset

```python
data = pd.read_csv('dataset.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
import plotly.express as px

px.line(data_frame = data, x='date', y='reference_evapotranspiration')
```

Let's create the above graph for 3 days instead of a year to visualise it clearly

```python
first_three_days = data.iloc[0:73]

first_three_days
```

```python
px.line(data_frame = first_three_days, x='date', y='reference_evapotranspiration')
```

From the above plot it's clear that the our target_variable value starts decreasing as soon as the sun sets and it's minimum throughout the whole night then it starts increasing with sunrise and is highest at around 16:00

## Plotting correlation matrix

```python
corr_mat=data.corr(method='pearson')
plt.figure(figsize=(20,10))
sns.heatmap(corr_mat,vmax=1,square=True,annot=True,cmap='cubehelix')
```

* from above correlation matrix we can say that, `maximum_atmospheric_pressure` and `minimum_atmospheric_pressure` are not relating much to our target feature i.e `reference_evatranspiration` so we can drop that value at some point from our dataset

* Also the features `average_relative_humidity`, `average_air_temperature`, `wind_speed` are the ones correalting much to our target feature, so let's visulise it more clearly 

## plot of `average_relative_humidity` and `reference_evapotranspiration`

```python
data_ar = data[['date', 'average_relative_humidity', 'reference_evapotranspiration']]

data_ar
```

```python
data_ar['reference_evapotranspiration'] = data_ar['reference_evapotranspiration']*1000
data_ar
```

```python
px.line(data_frame=data_ar[:73], x='date', y=data_ar.columns[1:])
```

We can see there is a nice relation between `average_relative_humidity` and `reference_evapotranspiration`, i.e. when one increases other decreases and vice versa

## plot of `average_air_temperature` and `reference_evapotranspiration`

```python
data_tr = data[['date', 'average_air_temperature', 'reference_evapotranspiration']]

data_tr
```

```python
data_tr['reference_evapotranspiration'] = data_tr['reference_evapotranspiration']*1000

data_tr
```

```python
px.line(data_frame=data_tr[:73], x='date', y=data_tr.columns[1:])
```

Here also we have a nice relation between both features we need to normalise the value of all feature to transform all values in same range
